### Dataset Splitting (Traing Set , Test Set)

```python
#splitting the dataset into input feature and the output
X = med_dataset.drop('prognosis', axis=1) # Removes the target column from the feature set
Y = med_dataset['prognosis'] # Stores the target labels separately
```

```python
X
```

```python
Y
```

```python
#encoding the target variable prognosis using scikit-learn’s LabelEncoder.

# Initialize the encoder
le = LabelEncoder()

# Fit on the target values
le.fit(Y)

# Transform the labels into numeric values
Y = le.transform(Y)
```

This converts categorical class labels (e.g., 'Flu', 'Cold', 'Covid') into integers (e.g., 0, 1, 2), which are easier for most machine learning models to work with.

```python
#dataset Splitting
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=20)
# 30% for testing, 70% for training
#X_train: training features (70%)
#X_test: testing features (30%)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

### Models Selection and Training

```python
#The reason for importing many models in this script is usually to perform a model comparison.
#that is, to train multiple classifiers on the same dataset and compare their performance.

# Create a dictionary to store models
models = {
    'SVC': SVC(kernel='linear'), #kernel='linear' means it assumes the data is linearly separable 
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42), #n_estimators=100 means it builds 100 trees.
    'GradientBoosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'KNeighbors': KNeighborsClassifier(n_neighbors=5),
    'MultinomialNB': MultinomialNB()
}
```

| Model     | Description |
|:----------|---------:|
| RF        | Ensemble of decision trees, robust and widely used.     |
| SVM       | 	Good for small to medium datasets with a clear margin of separation.   |
| GradientBoostingClassifier | 	Sequential boosting model, often more accurate. |
| KNeighborsClassifier | Instance-based, makes predictions based on closest neighbors. |
| MultinomialNB | Naive Bayes, good for text and count-based features. |


```python
# Loop through the models, train, test, and print results
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Test the model
    predictions = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, predictions)
    print(f"{model_name} Accuracy: {accuracy:.4f}")

    # Calculate confusion matrix
    cm = confusion_matrix(y_test, predictions)
    print(f"{model_name} Confusion Matrix:")
    print(np.array2string(cm, separator=', ')) #is used to neatly print the confusion matrix (cm) as a formatted string.
    print("\n" + "="*40 + "\n")
```

For a perfect classifier:<br>
All non-zero values are on the diagonal.<br>
The diagonal values (like 40, 43, 28...) are the correct predictions for each class.<br>
All off-diagonal values are 0 (i.e., no misclassifications).

##### Training using SVM Model
we have selected SVM because it the most sutiable for our dataset.

```python
# selecting svc
svc = SVC(kernel='linear')
svc.fit(X_train,y_train)
ypred = svc.predict(X_test)
accuracy_score(y_test,ypred)
```

##### Saving the model after training

```python
pickle.dump(svc,open('svc.pkl','wb'))
#This saves (dumps) the trained model object svc into a file called svc.pkl.
#svc: the trained model object you want to save.
#'svc.pkl': the filename where the model will be stored.
#'wb': write binary mode — required for storing objects in binary form.
# When you save model,You don't have to retrain the model every time you use it.
```

##### Loading model after saving

```python
# load model
svc = pickle.load(open('svc.pkl','rb'))
#You don't have to retrain the model every time you use it.
#'svc.pkl' is the name of the file where your model was saved earlier using pickle.dump.
#'rb' stands for "read binary" mode — because the model was saved in binary form.
```

### Loading Database and Linking it with Predicitions

```python
sym_des = pd.read_csv("Database/symtoms_df.csv")
precautions = pd.read_csv("Database/precautions_df.csv")
workout = pd.read_csv("Database/workout_df.csv")
description = pd.read_csv("Database/description.csv")
medications = pd.read_csv('Database/medications.csv')
diets = pd.read_csv("Database/diets.csv")
```

##### helper funciton
This function gathers related information for a given disease from database tables.<br>
This function takes a disease name dis and returns relevant descriptive information.

| Variable | Definition |
|--|--|
|desc| gets the disease description from description table.|
|pre| gets a list of four precaution steps for the disease.|
|med| gets a list of recommended medications.|
|die| gets diet recommendations.|
|wrkout| gets suggested workouts or physical activities.|

```python
def helper(disease):
    desc = description[description['Disease'] == disease]['Description']
    # go to the description table check if the value in Disease column = the predicted disease then return the Description.
    desc = " ".join([w for w in desc]) #convert a pandas Series (desc) into a single string. 
    #his list comprehension iterates over the Series and collects each string value into a list.
    
    pre = precautions[precautions['Disease'] == disease][['Precaution_1', 'Precaution_2', 'Precaution_3', 'Precaution_4']]
    pre = [col for col in pre.values] #produces a list with a single element — the row itself — like this
                                      #  ['Drink more water', 'Take rest', 'Avoid junk food', 'Consult doctor']

    med = medications[medications['Disease'] == disease]['Medication']
    med = [med for med in med.values]

    die = diets[diets['Disease'] == disease]['Diet']
    die = [die for die in die.values]

    wrkout = workout[workout['disease'] == disease] ['workout']


    return desc,pre,med,die,wrkout

```
